```python
data = read_csv("C:\Users\SamirM1\Desktop\competency\data.csv")

acomplete = [d for d in data.Percent if (d == 100)]  
azero = [d for d in data.Percent if (d == 0)]  
aprogress = [d for d in data.Percent if (d > 0 and d < 100)]  
```

```python
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),  
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),  
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),  
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),  
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]  
    
for i in range(len(tableau20)):  
    r, g, b = tableau20[i]  
    tableau20[i] = (r / 255., g / 255., b / 255.)  
    
figure(1, figsize=(8,8))
ax = axes([0.1, 0.1, 0.8, 0.8])
color = tableau20
labels = 'In Progress', 'Complete', 'Zero'
fracs = [len(aprogress), len(acomplete), len(azero)]
explode=(0, 0.05, 0)

pie(fracs, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, colors=color)
                
title('Completion Percentage', bbox={'facecolor':'1.0', 'pad':10})

show()
```
